```python
import pandas as pd
import altair as alt
DF = pd.read_csv('ClenData.csv')
DF = DF.drop(columns=['Unnamed: 0'])
DF
```

```python
# department
selection = alt.selection(type = "multi", fields = ["Department"], on = "mouseover", nearest=True)
alt.Chart(DF).mark_bar().encode(
    x = "Department",
    y = "mean(Q6-How Satisfied You are with your Student Life ?):Q", 
    opacity = alt.condition(selection, alt.value(1), alt.value(.2))
).add_selection(selection)
```

```python
selection = alt.selection(type="multi", fields=["Housing Type"])
# this creates a container for our two vizualizations
base = alt.Chart(DF).properties(width = 500, height = 250)

# to create an overview chart 
overview = alt.Chart(DF).mark_bar().encode(
    x = "Housing Type",
    y = "mean(Q6-How Satisfied You are with your Student Life ?):Q",
    color = alt.condition(selection, alt.value("lightblue"), alt.value("lightgrey"))
).add_selection(selection).properties(height=250)

# to create a detailed chart
detail = base.mark_bar().encode(
    x = "Department",
    y = "mean(Q6-How Satisfied You are with your Student Life ?):Q",
).transform_filter(selection).properties(height = 250)
# transform_filter(selection) says filter out all the data that doesn't match my selection

overview | detail
```
